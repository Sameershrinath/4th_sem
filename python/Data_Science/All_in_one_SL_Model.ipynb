```python
# üìö Importing required libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, mean_squared_error, r2_score

# üìà Models
from sklearn.linear_model import LinearRegression          # Linear Regression
from sklearn.linear_model import LogisticRegression        # Logistic Regression
from sklearn.svm import SVC                                # Support Vector Machine
from sklearn.tree import DecisionTreeClassifier            # Decision Tree
from sklearn.ensemble import RandomForestClassifier        # Random Forest

# üéØ Load datasets
from sklearn.datasets import load_iris                     # For classification
from sklearn.datasets import fetch_california_housing      # For regression

# üß™ Classification Task (Iris)
iris = load_iris()
X_cls = iris.data
y_cls = iris.target

# üßÆ Regression Task (California Housing)
california = fetch_california_housing()
X_reg = california.data
y_reg = california.target

# ‚úÇÔ∏è Split datasets
X_train_cls, X_test_cls, y_train_cls, y_test_cls = train_test_split(X_cls, y_cls, test_size=0.2, random_state=42)
X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(X_reg, y_reg, test_size=0.2, random_state=42)

# -------------------------------
# üîπ Linear Regression (Regression)
lr = LinearRegression()
lr.fit(X_train_reg, y_train_reg)
y_pred_lr = lr.predict(X_test_reg)

print("üîπ Linear Regression:")
print("MSE:", mean_squared_error(y_test_reg, y_pred_lr))
print("R¬≤ Score:", r2_score(y_test_reg, y_pred_lr))
print()

# -------------------------------
# üîπ Logistic Regression (Classification)
log_reg = LogisticRegression(max_iter=200)
log_reg.fit(X_train_cls, y_train_cls)
y_pred_log = log_reg.predict(X_test_cls)

print("üîπ Logistic Regression:")
print("Accuracy:", accuracy_score(y_test_cls, y_pred_log))
print("Report:\n", classification_report(y_test_cls, y_pred_log))
print()

# -------------------------------
# üîπ SVM (Classification)
svm = SVC(kernel='rbf', gamma='scale')
svm.fit(X_train_cls, y_train_cls)
y_pred_svm = svm.predict(X_test_cls)

print("üîπ Support Vector Machine (SVM):")
print("Accuracy:", accuracy_score(y_test_cls, y_pred_svm))
print("Report:\n", classification_report(y_test_cls, y_pred_svm))
print()

# -------------------------------
# üîπ Decision Tree (Classification)
dt = DecisionTreeClassifier()
dt.fit(X_train_cls, y_train_cls)
y_pred_dt = dt.predict(X_test_cls)

print("üîπ Decision Tree:")
print("Accuracy:", accuracy_score(y_test_cls, y_pred_dt))
print("Report:\n", classification_report(y_test_cls, y_pred_dt))
print()

# -------------------------------
# üîπ Random Forest (Classification)
rf = RandomForestClassifier(n_estimators=100)
rf.fit(X_train_cls, y_train_cls)
y_pred_rf = rf.predict(X_test_cls)

print("üîπ Random Forest:")
print("Accuracy:", accuracy_score(y_test_cls, y_pred_rf))
print("Report:\n", classification_report(y_test_cls, y_pred_rf))
print()
```
