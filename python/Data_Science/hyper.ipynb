```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

x=np.random.randint(25,75,100).reshape(-1,1)
y=4*x+10

x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=2,test_size=0.1)

model=LinearRegression()
model.fit(x_train,y_train)
```

```python
model.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error

mean_squared_error(x_test,model.predict(x_test))
```

```python
from  sklearn.linear_model import LogisticRegression

x1 = np.random.randint(200, 500, 1000)
x2 = np.random.randint(50, 100, 1000)
x = pd.DataFrame({
    'feature1': x1,
    'feature2': x2
})
y=[ np.random.choice([0,1]) for i in range(0,1000) ]

x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=2,test_size=0.2)

model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
model.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import f1_score


y_pred = model.predict(x_test)
cm = confusion_matrix(y_test, y_pred)
f1_scorel = f1_score(y_test, y_pred)

print(cm)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.heatmap(cm,annot=True)
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate random x values
x = np.random.uniform(-10, 10, 500)

# Step 2: Define class boundary: y = 1 if x^2 > 25 else 0
threshold = 25
y = np.where(x**2 > threshold, 1, 0)

# Step 3: Visualization
plt.scatter(x, y)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

# Load a simple classification dataset
X, y = datasets.make_classification(n_samples=200, n_features=2, n_redundant=0,
                                     n_clusters_per_class=1, random_state=42)

# Standardize features
scaler = StandardScaler()
X = scaler.fit_transform(X)

# Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Define values of C and gamma
C_values = [0.1, 1, 10]
gamma_values = [0.1, 1, 10]

# Plotting
plt.figure(figsize=(12, 10))

for i, C in enumerate(C_values):
    for j, gamma in enumerate(gamma_values):
        model = SVC(kernel='rbf', C=C, gamma=gamma)
        model.fit(X_train, y_train)

        # Plot
        plt.subplot(len(C_values), len(gamma_values), i * len(gamma_values) + j + 1)
        plt.title(f'C={C}, gamma={gamma}')

        # Plot decision boundary
        x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
        y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
        xx, yy = np.meshgrid(np.linspace(x_min, x_max, 500),
                             np.linspace(y_min, y_max, 500))
        Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
        Z = Z.reshape(xx.shape)
        plt.contourf(xx, yy, Z, alpha=0.3, cmap=plt.cm.coolwarm)
        plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train, cmap=plt.cm.coolwarm, s=10)
        plt.xlabel('Feature 1')
        plt.ylabel('Feature 2')

plt.tight_layout()
plt.show()
```
