```python
from sklearn.preprocessing import OneHotEncoder,LabelEncoder
df3 = pd.DataFrame({'City': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix', 'Philadelphia', 'San Antonio', 'San Diego', 'Dallas', 'San Jose']})
```

```python
onehotencoder = OneHotEncoder()
df3_encoded = onehotencoder.fit_transform(df3[['City']]).toarray()

# Create a DataFrame with the encoded data
df3_encoded_df = pd.DataFrame(df3_encoded, columns=onehotencoder.get_feature_names_out(['City']))
df3_encoded_df
```
