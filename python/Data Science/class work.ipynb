```python
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt

# Generate random data near 50
np.random.seed(0)  
BHK = np.random.randint(45, 55, size=100)
price = BHK * 1000 + np.random.normal(0, 1000, size=100)  # Adding some noise

# Reshape the data
X = np.array(BHK).reshape(-1, 1)
y = np.array(price)

# Create and fit the model
model = LinearRegression()
model.fit(X, y)

# Make predictions
y_pred = model.predict(X)

# Plot the results
plt.scatter(BHK, price, color='blue', label='Actual',s=2)
plt.plot(BHK, y_pred, color='red', label='Predicted')
plt.xlabel('BHK')
plt.ylabel('Price')
plt.title('Linear Regression: Price vs BHK')
plt.legend()
plt.show()
```

```python
# Assumptions of Linear Regression

assumptions = """
1. Linearity: The relationship between the independent variable (BHK) and the dependent variable (price) is linear.
2. Independence: The residuals (errors) are independent. This means that the error terms are not correlated with each other.
3. Homoscedasticity: The residuals have constant variance at every level of the independent variable.
4. Normality: The residuals of the model are normally distributed.
5. No Multicollinearity: The independent variables are not highly correlated with each other (not applicable in simple linear regression with one independent variable).
"""

print(assumptions)
```
