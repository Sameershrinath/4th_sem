```python
condition = (df['content_rating'] == 'PG') & (df['genre'] == 'Action')
condition.iloc[:20]
```

```python

condition = (df['content_rating'] == 'PG') & ( (df['genre'] == 'Action') | (df['genre'] == 'Crime') )
df.loc[condition]
```

```python
df.loc[(df['content_rating'] == 'PG') & 
       ( (df['genre'] == 'Action') | 
        (df['genre'] == 'Crime') )]
```

```python
df.loc[(df['content_rating'] == 'R') & 
       ( (df['genre'] == 'Comedy') | 
        (df['genre'] == 'Crime') )]
```

```python
df.loc[(df['content_rating'] == 'R') & 
       ( (df['genre'] == 'Action') |
        (df['genre'] == 'Crime') ) &
      (df['duration'] <= 180)]
```

```python
df.loc[(df['content_rating'] == 'R') & 
       ( (df['genre'] == 'Action') | 
        (df['genre'] == 'Crime') ) &
      (df['duration'] <= 180) &
      (df['duration'] >= 120)]
```

```python
df.loc[(df['content_rating'] == 'R') & 
       ( (df['genre'] == 'Action') | 
        (df['genre'] == 'Crime') ) &
      (df['duration'] <= 180) &
      (df['duration'] >= 150)]
```

```python
df.loc[(df['content_rating'] == 'R') & 
       ( (df['genre'] == 'Action') | 
        (df['genre'] == 'Crime') ) &
      (df['duration'] <= 180) &
      (df['duration'] >= 150) &
      (df['star_rating'] >= 8)]
```

```python
condition = (df['content_rating'] == 'R') & ((df['genre'] == 'Action') | (df['genre'] == 'Crime')) & (df['duration'] <= 180) & (df['duration'] >= 150) & (df['star_rating'] >= 8)
        
df.loc[condition]
```

```python
condition = (df['content_rating'] == 'R') & ((df['genre'] == 'Action') | (df['genre'] == 'Crime')) & (df['duration'] <= 180) & (df['duration'] >= 150) & (df['star_rating'] >= 8)
        
df.loc[condition, ['title', 'actors_list']]
```

```python
df.loc[(df['genre'] == 'Crime') |
      (df['genre'] == 'Action') |
      (df['genre'] == 'Drama')]
```
