# 4. Create Excel File

```python
movie_df = pd.read_csv("movies_data.csv")
movie_df
```

```python
movie_rating_df = movie_df[['title', 'star_rating', 'content_rating']]
# movie_rating_df
```

```python
movie_genre_df = movie_df[['title', 'genre', 'duration']]
# movie_genre_df
```

```python
movie_actor_df = movie_df[['title', 'actors_list']]
# movie_actor_df
```

```python
with pd.ExcelWriter("movies_details.xlsx") as movie:
    movie_rating_df.to_excel(movie,sheet_name="movie ratings",index=False)
    movie_genre_df.to_excel(movie,sheet_name="movie genre",index=False)
    movie_actor_df.to_excel(movie,sheet_name="movie actor",index=False)
```

```python
Pandas provide a function called xlsxwriter for this purpose. 
ExcelWriter() is a class that allows you to write DataFrame objects into Microsoft Excel sheets.
Text, numbers, strings, and formulas can all be written using ExcelWriter().
It can also be used on several worksheets.
```
