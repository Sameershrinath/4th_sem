# 2. sort_values()

```python
df
```

```python
df.sort_values('Weight in Kgs')  # Ascending order
```

```python
df.sort_values('Weight in Kgs', ascending=False)  # Descending order
```

```python
df.sort_values('First Name')  # Ascending order
```

```python
df.sort_values('City')  # Ascending order
```

```python
df.sort_values('Weight in Kgs')  # Ascending order
```

```python
df.sort_values(['Weight in Kgs', 'City'])  # Ascending order
```

```python
df
```

```python
df.sort_values(['Salary'], ascending=False)
```

```python
df[['Salary']].sort_values(['Salary'], ascending=False)
```

```python
highest_salary = df[['Salary']].sort_values(['Salary'], ascending=False).iloc[0,0]
print("Higest Salary -->", highest_salary)
```

```python
# Second higest salary
df[['Salary']].sort_values(['Salary'], ascending=False).iloc[1,0]
```

```python
# Get maximum experience in company
df[['Age in Company']].sort_values(['Age in Company'], ascending=False).iloc[0,0]
```

```python
df2 = df.sort_values(['Salary'], ascending=False)
df2
```

# 3. reset_index()

```python
df2.reset_index()
```

```python
df2.reset_index(drop= True)
```
