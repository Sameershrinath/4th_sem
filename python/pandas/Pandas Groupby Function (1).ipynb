### Find first 5 movies 'title's only with content rating R and genre is comedy, sorted with star_rating in ascending order in a list

```python
movies_list = df.groupby(['genre', 'content_rating']).get_group(("Comedy", "R")).sort_values("star_rating").head()['title'].tolist()
movies_list
```

```python
movies_list1 = df.groupby(['genre', 'content_rating']).get_group(("Comedy", "R")).sort_values("star_rating")
movies_list1
```
