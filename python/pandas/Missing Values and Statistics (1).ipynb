```python
import pandas as pd
import numpy as np
```

```python
1. Detect Missing Values:
    1. df.isna()
    2. df.isna().sum()
    3. df.isna().mean() * 100
    4. df.isnull()
    
2. Drop Null Values:
    1. df.dropna(thresh=)  # At least thresh value is non null  # by default axis=0
        df.dropna(thresh=cnt, axis=1) 
    2. df.drop('Row_name', axis=0, inplace= True)
        df.drop([row1, row2, row3], inplace=True)
        
        df.drop('col_name', axis=1, inplace= True)
        df.drop([col1, col2, col3], axis=1, inplace=True)
        
3. Impute Missing values (mean, median, mode):
    df.fillna(value, inplace=True)
    
    statistic methods:
        1.df['col_name'].mean()
        2.df['col_name'].median()
        3.df['col_name'].mode()[0]
        
mean >> If there is no outlier
median >> If there is outlier in data
mode >> When data is in categorical form
```

```python
import pandas as pd

dict1 = {'a':[10],'b':[20], 'c':[50],'d':[40],'e':[80]}
df2 = pd.DataFrame(dict1)
df2
```

### Method-1

```python
# We have to use append function
df = df.append(df2)
df
```

```python
# We can append dataframe inside dataframe
# We are getting appendded index of datframe at last.
# If we have to ignore this index, we can give ignore_index = True

df = pd.DataFrame(array, columns=list('abcde'))

dict1 = {'a':[10],'b':20, 'c':50,'d':40,'e':80}
df2 = pd.DataFrame(dict1, index=[0])

df = df.append(df2, ignore_index=True)
df
```

### Method-2

```python
# By using concat function

df1 = pd.DataFrame(array, columns=list('abcde'))

dict1 = {'a':[10],'b':20, 'c':50,'d':40,'e':80}
df2 = pd.DataFrame(dict1, index=[0])

df = pd.concat([df1,df2], ignore_index=True)  # We are goin to see this pandas function concat in detail
df
```
