```python
movies_df = pd.read_csv("movies_data.csv")
movies_df
```

```python
movies_df['title'] = movies_df['title'].apply(lambda x : x.upper())
movies_df
```

```python
import re

def get_movie_title(x):
    movies_name = re.sub('[^a-zA-Z]+', " ", x)
    return movies_name

x = "MASTER AND COMMANDER: THE FAR SIDE OF THE WORLD"
get_movie_title(x)
```

```python
movies_df['title'] = movies_df['title'].apply(get_movie_title)
movies_df
```

```python
x = "[u'Tim Robbins', u'Morgan Freeman', u'Bob Gunton']"
x = x.replace("u'", "")   # replace u'

x = re.sub('[^a-zA-Z,]', " ", x)  # Removing everything otherr than alphabets and ,

x = [i.strip() for i in x.split(",")]  # Getting list by removing leading and trailing spaces
x
```

```python
def get_actor_list(x):
    x = x.replace("u'", "")  # replacing u'
    
    actor = re.sub('[^a-zA-Z,]', " ", x)
    
    actors_list = [i.strip() for i in actor.split(",")]
    
    return actors_list

x = "[u'Tim Robbins', u'Morgan Freeman', u'Bob Gunton']"
get_actor_list(x)    
```

```python
movies_df['actors_list'] = movies_df['actors_list'].apply(get_actor_list)
movies_df
```
